```python
import numpy as np
import matplotlib.pyplot as plt

#dimensiones a analizar
dimensiones = [10, 50, 100, 500, 1000, 2000, 5000]

for dim in dimensiones:
    #cargar los datos de distancia y frecuencia desde el archivo .txt
    filename = f"distancias_dim_{dim}.txt"
    data = np.loadtxt(filename)

    #etraer las distancias del archivo de datos
    distancias = data.flatten()

    #hstograma de las distancias
    plt.figure(figsize=(7, 3))
    plt.hist(distancias, bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histograma de Distancias - Dimensión {dim}')
    plt.xlabel('Distancia')
    plt.ylabel('Frecuencia')
    plt.grid(True)

    #mostrar el histograma
    plt.show()

```
